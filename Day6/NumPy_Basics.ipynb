# ------------------------------------------------------
# ðŸ§© 1. Importing NumPy and Checking Version
# ------------------------------------------------------

```python
import numpy as np

print("NumPy version:", np.__version__)
```

# ------------------------------------------------------
# ðŸ§© 2. Creating Arrays
# ------------------------------------------------------

```python

# 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr1)

# 2D array
arr2 = np.array([[10, 20, 30], [40, 50, 60]])
print("\n2D Array:\n", arr2)

# Array of zeros
zeros = np.zeros((3, 3))
print("\nZeros Array:\n", zeros)

# Array of ones
ones = np.ones((2, 4))
print("\nOnes Array:\n", ones)

# Range of numbers
range_arr = np.arange(0, 10, 2)  # start=0, stop=10, step=2
print("\nRange Array:", range_arr)

# Linearly spaced numbers between 0 and 1
linear_arr = np.linspace(0, 1, 5)
print("\nLinearly spaced Array:", linear_arr)

# Random integers between 1 and 10 in 3x3 matrix
rand_arr = np.random.randint(1, 10, (3, 3))
print("\nRandom 3x3 Array:\n", rand_arr)
```

# ------------------------------------------------------
# ðŸ§© 3. Array Attributes
# ------------------------------------------------------

```python
print("\nArray Shape:", rand_arr.shape)
print("Array Size:", rand_arr.size)
print("Array Dimensions:", rand_arr.ndim)
print("Array Data Type:", rand_arr.dtype)
```

# ------------------------------------------------------
# ðŸ§© 4. Indexing & Slicing
# ------------------------------------------------------

```python
a = np.arange(10)
print("\nOriginal Array:", a)
print("Slice [2:7]:", a[2:7])
print("Every 2nd element:", a[::2])
print("Reversed Array:", a[::-1])

matrix = np.arange(12).reshape(3, 4)
print("\nMatrix:\n", matrix)
print("Element at (1,2):", matrix[1, 2])
print("Second column:", matrix[:, 1])
print("First row:", matrix[0, :])
```

# ------------------------------------------------------
# ðŸ§© 5. Array Mathematics
# ------------------------------------------------------

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print("\nAddition:", x + y)
print("Subtraction:", y - x)
print("Multiplication:", x * y)
print("Division:", y / x)
print("Mean of x:", x.mean())
print("Dot Product:", np.dot(x, y))
```

# ------------------------------------------------------
# ðŸ§© 6. Reshape & Transpose
# ------------------------------------------------------

```python
mat = np.arange(6).reshape(2, 3)
print("\nOriginal Matrix:\n", mat)
print("Transposed Matrix:\n", mat.T)
```

# ------------------------------------------------------
# ðŸ§© 7. Broadcasting
# ------------------------------------------------------

```python
# Broadcasting allows operations on arrays of different shapes
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([10, 20, 30])
print("\nBroadcasting Result:\n", a + b)
```

# ------------------------------------------------------
# ðŸ§© 8. Statistics on Arrays
# ------------------------------------------------------

```python
stats_arr = np.random.randint(10, 100, (4, 4))
print("\nArray for Stats:\n", stats_arr)
print("Sum:", stats_arr.sum())
print("Mean:", stats_arr.mean())
print("Min:", stats_arr.min())
print("Max:", stats_arr.max())
print("Standard Deviation:", stats_arr.std())
```

# ------------------------------------------------------
# ðŸ§© 9. Mini Project: Matrix Playground
# ------------------------------------------------------

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create random 4x4 matrix
matrix = np.random.randint(1, 50, (4, 4))
print("\nMatrix Playground:\n", matrix)

# Step 2: Compute basic stats
print("Sum of all elements:", np.sum(matrix))
print("Mean of all elements:", np.mean(matrix))
print("Transpose:\n", matrix.T)

# Step 3: Dot product of matrix with its transpose
dot_result = np.dot(matrix, matrix.T)
print("\nDot Product (Matrix x Transpose):\n", dot_result)

# Step 4: Visualize as heatmap
plt.imshow(matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Matrix Heatmap")
plt.show()

sns.heatmap(matrix, annot=True, fmt="d", cmap='coolwarm')
plt.title("Matrix Heatmap with Values")
plt.show()
```
