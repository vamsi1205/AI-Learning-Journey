```python
# ---------------------------------------------------------------
# üéì Student Performance Analyzer - NumPy Mini Project
# Author: Vamsi
# Goal: Analyze student marks using NumPy fundamentals
# ---------------------------------------------------------------

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------------
# üß© Step 1: Create Random Marks Data
# ---------------------------------------------------------------
np.random.seed(42)  # reproducibility

# 10 students, 5 subjects (marks between 35 and 100)
marks = np.random.randint(35, 100, (10, 5))

# Label arrays for clarity
students = np.arange(1, 11)
subjects = np.array(["Math", "Science", "English", "History", "Computers"])

print("üéØ Student Marks Matrix:\n")
print(marks)

# ---------------------------------------------------------------
# üßÆ Step 2: Calculate Totals and Averages
# ---------------------------------------------------------------
totals = marks.sum(axis=1)           # total per student
avg_per_student = marks.mean(axis=1) # average per student
avg_per_subject = marks.mean(axis=0) # average per subject

print("\nüìä Average Marks per Subject:")
for s, avg in zip(subjects, avg_per_subject):
    print(f"{s:<10}: {avg:.2f}")

# ---------------------------------------------------------------
# üèÜ Step 3: Identify Topper and Weakest Subject
# ---------------------------------------------------------------
topper_index = np.argmax(totals)
hardest_subject_index = np.argmin(avg_per_subject)

print(f"\nüèÖ Topper: Student {students[topper_index]} with total {totals[topper_index]}")
print(f"üìâ Hardest Subject: {subjects[hardest_subject_index]} with avg {avg_per_subject[hardest_subject_index]:.2f}")

# ---------------------------------------------------------------
# ‚ö†Ô∏è Step 4: Find Students Who Failed in Any Subject (<40)
# ---------------------------------------------------------------
failed_subjects = np.sum(marks < 40, axis=1)  # count of failed subjects per student
failed_students_mask = failed_subjects > 0    # boolean mask

print("\n‚ùå Students Who Failed in At Least One Subject:")
for sid, fails in zip(students, failed_subjects):
    if fails > 0:
        print(f"Student {sid} failed in {fails} subject(s)")

# ---------------------------------------------------------------
# üéÅ Step 5: Add Bonus Marks (+5) to Students Who Failed At Least Once
# ---------------------------------------------------------------
# Add 5 marks to all subjects for students who failed at least once
marks[failed_students_mask] += 5

# Ensure marks do not exceed 100
marks = np.clip(marks, 0, 100)

print("\n‚úÖ Bonus Marks Added for Failed Students (max capped at 100):\n")
print(marks)

# ---------------------------------------------------------------
# üîÅ Step 6: Recalculate Totals and Averages After Bonus
# ---------------------------------------------------------------
new_totals = marks.sum(axis=1)
new_avg_per_student = marks.mean(axis=1)
new_avg_per_subject = marks.mean(axis=0)

print("\nüìà New Average Marks per Subject (After Bonus):")
for s, avg in zip(subjects, new_avg_per_subject):
    print(f"{s:<10}: {avg:.2f}")

# ---------------------------------------------------------------
# üìä Step 7: Visualization
# ---------------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.bar(subjects, new_avg_per_subject, color='skyblue', edgecolor='black')
plt.title("Average Marks per Subject (After Bonus)")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.ylim(0, 100)
plt.show()

# ---------------------------------------------------------------
# üßæ Step 8: Summary Report
# ---------------------------------------------------------------
records = [
    {
        "Student": int(sid),
        "Marks": row.tolist(),
        "Total": int(total),
        "FailedSubjects": int(fail),
        "NewTotal": int(new_total)
    }
    for sid, row, total, fail, new_total in zip(students, marks, totals, failed_subjects, new_totals)
]

print("\nüìã Final Student Summary:")
for record in records:
    print(record)
```
