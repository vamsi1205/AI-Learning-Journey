```python
# ---------------------------------------------------------------
# ðŸŽ² NumPy Mini Project - Dice Roll Simulator
# Author: Vamsi
# Goal: Simulate dice rolls, calculate probabilities, visualize results
# ---------------------------------------------------------------

# Importing necessary libraries
import numpy as np               # For numerical computation and random generation
import matplotlib.pyplot as plt  # For data visualization (bar charts, plots)

# ---------------------------------------------------------------
# ðŸ§© Step 1: Set Random Seed
# ---------------------------------------------------------------
# Setting a seed value ensures reproducibility â€” meaning
# every time you run this code, youâ€™ll get the same random numbers.
np.random.seed(7)

# ---------------------------------------------------------------
# ðŸŽ¯ Step 2: Simulate Dice Rolls
# ---------------------------------------------------------------
# np.random.randint(low, high, size)
# Generates random integers between 'low' (inclusive) and 'high' (exclusive)
# Here: numbers between 1 and 6 (6-sided die)
# Simulate 10,000 dice rolls
rolls = np.random.randint(1, 7, 10000)

# ---------------------------------------------------------------
# ðŸ“Š Step 3: Count Occurrences of Each Face
# ---------------------------------------------------------------
# np.unique() finds all unique values in an array
# return_counts=True returns how many times each unique value appears
values, counts = np.unique(rolls, return_counts=True)

# ---------------------------------------------------------------
# ðŸ”¢ Step 4: Calculate Probabilities
# ---------------------------------------------------------------
# Probability = count of each face / total number of rolls
probabilities = counts / len(rolls)

# ---------------------------------------------------------------
# ðŸ§® Step 5: Print Results in a Table Format
# ---------------------------------------------------------------
print("Face | Count | Probability")
print("----------------------------")
for v, c, p in zip(values, counts, probabilities):
    # f-string formatting for aligned and readable output
    print(f" {v:4d} | {c:5d} | {p:.3f}")

# ---------------------------------------------------------------
# ðŸŽ¨ Step 6: Visualize the Probability Distribution
# ---------------------------------------------------------------
# Create a bar chart of dice faces vs probabilities
plt.bar(values,               # X-axis: dice faces (1â€“6)
        probabilities,        # Y-axis: probability for each face
        color='orange',       # Bar color
        edgecolor='black')    # Border color for better visibility

# Add title and labels
plt.title("Dice Roll Probability Distribution")
plt.xlabel("Die Face")
plt.ylabel("Probability")

# Display the plot window
plt.show()
```
