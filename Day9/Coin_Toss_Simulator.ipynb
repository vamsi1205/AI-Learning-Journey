```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
N = 1000
tosses = np.random.randint(0, 2, N)   # 0 = Heads, 1 = Tails

# Calculate cumulative probability of heads
heads_cum = np.cumsum(tosses == 0)
prob_over_time = heads_cum / np.arange(1, N+1)

plt.plot(prob_over_time, color='purple')
plt.axhline(0.5, color='red', linestyle='--', label='Expected Probability (0.5)')
plt.title("Probability Convergence of Coin Toss")
plt.xlabel("Number of Tosses")
plt.ylabel("P(Heads)")
plt.legend()
plt.show()
```
