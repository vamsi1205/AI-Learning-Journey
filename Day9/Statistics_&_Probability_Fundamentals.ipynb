```python
import numpy as np

data = np.array([5, 8, 12, 20, 7, 9, 14, 10, 9, 8])

mean = np.mean(data)
median = np.median(data)

# NumPy doesn’t have direct mode; use Python’s statistics or SciPy
from scipy import stats

mode = stats.mode(data, keepdims=True)[0][0]

print(f"Mean: {mean:.2f}")
print(f"Median: {median}")
print(f"Mode: {mode}")

variance = np.var(data)
std_dev = np.std(data)
data_range = np.ptp(data)  # max - min

print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Range: {data_range}")

# Probability of rolling each number from 1 to 6
outcomes = np.arange(1, 7)
prob_each = np.full(6, 1 / 6)
print("Each outcome has probability:", prob_each)

# Simulate 1000 coin tosses (0=Heads, 1=Tails)
np.random.seed(42)
tosses = np.random.randint(0, 2, 1000)

heads_prob = np.sum(tosses == 0) / len(tosses)
tails_prob = np.sum(tosses == 1) / len(tosses)

print(f"Heads Probability: {heads_prob:.3f}")
print(f"Tails Probability: {tails_prob:.3f}")


import matplotlib.pyplot as plt

plt.hist(data, bins=5, color="orange", edgecolor="black")
plt.title("Data Distribution")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```
